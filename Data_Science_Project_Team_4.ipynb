# Analyze the relationship between budget, revenue, and popularity to determine factors that contribute to a movie's success. - JOSHUA

The most objective metric to determine whether a movie was a success or not is the profit of the movie. Profit could be defined as revenue - budget.

A few considerations: Is a profit of $1 million the same if the budget was $1 million versus a budget that was $100 million? No. In the first case the movie would be hailed as a success and likely a failure in the italicized text second case. So what we care about is not only the quantitative profit, but the percentage of profit over the budget, aka the profit margin.

Also it doesn't make sense to predict just profit using budget and revenue, because its obvious that the higher the revenue and lower the budget at the same time, the higher the profit since profit is derived from the two.

profit margin = (revenue - budget) / budget * 100%

So we'll perform an analysis to determine what features influenced the profit margin of the movies.

First we need to get the profit and profit margin for all the movies:

```python
# Create the "profit" column
movie_df['profit'] = movie_df['revenue'] - movie_df['budget']

# Create the "profit margin" column
movie_df['profit_margin'] = (movie_df['revenue'] - movie_df['budget']) / movie_df['budget']
# print(movie_df[['title', 'revenue', 'budget', 'profit_margin']].head())
print(movie_df[['title', 'revenue', 'budget', 'profit', 'profit_margin']].head())
```

Let's run a linear regression test.

Null hypothesis: 'budget', 'revenue', 'vote_average', 'popularity' have no significant effect on the profit margin of the movie.

Alternative hypothesis: 'budget', 'revenue', 'vote_average', 'popularity' have a significant effect on the profit margin of the movie.

First we need to clean up the data before we can run the linear regression test:

Let's remove all movies that are not yet released, since their metrics can only be made up rather than be historical data.

```python
# Filter the DataFrame to keep only rows where the "status" is "Released"
movie_df = movie_df[movie_df['status'] == 'Released']
```

```python
# Generate summary statistics
summary_stats = movie_df[['title', 'revenue', 'budget', 'vote_average', 'popularity', 'profit', 'profit_margin']].describe(include='all')
print(summary_stats)
```

Its clear that we need to fix the data up. The summary statistics are showing that we have max values unreasonably high numbers, which can only mean that they are bad data and we need to remove them. Also we have NAN values present which we need to get rid of

```python
# Define thresholds for reasonable values
revenue_threshold = 3_000_000_000  # 3 billion. I know that Avatar is highest grossing of all time, so that's our max value
budget_threshold = 500_000_000      # 500 million. I know that Avatar 2 and Star Wars Force Awakens are the most expensive movies ever made, so we can use that as our cap
profit_threshold = 3_000_000_000    # 3 billion. Using Avatar 2 revenue upper bound. Naive figure, but need to pick something
# profit_margin_threshold = 1000
# vote_average_threshold = 10          # Rating scale is 1-10
# popularity_threshold = 250

# Remove rows with unreasonably high values
movie_df = movie_df[
    (movie_df['revenue'] > 1000) &  #remove revenue of 0 or less
    (movie_df['revenue'] <= revenue_threshold) &
    (movie_df['budget'] <= budget_threshold) &
    (movie_df['profit'] <= profit_threshold)
    # (movie_df['profit_margin'] <= profit_margin_threshold)
    # (movie_df['vote_average'] <= vote_average_threshold) &
    # (movie_df['popularity'] <= popularity_threshold)
]

# Remove rows with NaN in key columns
movie_df = movie_df.dropna(subset=['title', 'revenue', 'budget', 'profit', 'profit_margin', 'vote_average', 'popularity'])

# Remove any remaining NaN values in other columns of interest
# This line ensures we remove rows with NaNs in any specified columns
movie_df = movie_df[['title', 'revenue', 'budget', 'profit', 'profit_margin', 'vote_average', 'popularity']].dropna()

# Check if there are still any NaN values
print("NaN values in DataFrame after cleaning:", movie_df.isna().sum())

# Print summary statistics for the specified columns
summary_stats = movie_df.describe(include='all')
print(summary_stats)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm

# Select the explanatory variables
X = movie_df[['budget', 'revenue', 'vote_average', 'popularity']]

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data
X_scaled = scaler.fit_transform(X)

# Convert the scaled array back to a DataFrame, ensuring the index is preserved
X_scaled = pd.DataFrame(X_scaled, columns=['budget', 'revenue', 'vote_average', 'popularity'], index=X.index)

# Add a constant term for the intercept
X_scaled = sm.add_constant(X_scaled)

# Dependent variable (make sure the index aligns)
y = movie_df['profit_margin']

# Ensure y has the same index as X_scaled
y = y[X_scaled.index]  # This will filter y to match the indices of X_scaled

# Fit the linear regression model
model = sm.OLS(y, X_scaled).fit()
print(model.summary())
```

Interpreting the results:

This is as low of an R-squared (and equally low adjusted R-squared) value as you can get. This means the model does a very poor job of explaining the variance in the profit margin. So there is weak explanatory power of the model regarding the profit margin.

The Prob F statistic very low at a statistically significant level, indicating that at least one of the predictor variables is a significant factor in predicting the profit margin.

The budget, revenue, and vote_average all have p-values that are statistically significant, so they have an effect on the profit margin.

A one standard deviation increase in revenue is associated with an increase of approximately 6.3663 units in the profit margin. So higher revenue leads to an increase in profit margin.

For each 1 standard deviation increase in the scaled budget, the profit margin is expected to decrease by approximately -7.3020 units. So its a negative relationship, higher budgets lead to lower profit margins.

A one unit increase in vote_average is associated with a decrease of approximately -2.3191 units in profit margin. This indicates that, counterintuitively, as a movie becomes more more highly rated, the profit margin might decrease.

What does revenue and budget coefficient mean in common sense english? If I were to increase the revenue or budget of a movie by $1 million, how is the profit margin predicted to change?

The revenue and budget were scaled to perform the linear regression, so we need to do some unscaling first for them before we can figure out their predicted impact on profit_margin.

```python
# Calculate means and standard deviations for the original variables
budget_mean = movie_df['budget'].mean()
budget_std = movie_df['budget'].std()
revenue_mean = movie_df['revenue'].mean()
revenue_std = movie_df['revenue'].std()

# Changes
change_in_revenue = 1_000_000
change_in_budget = 1_000_000

# Coefficients from the model
coef_budget = model.params['budget']
coef_revenue = model.params['revenue']

# Calculate predicted changes in profit margin
predicted_change_profit_margin_revenue = coef_revenue * (change_in_revenue / revenue_std)
predicted_change_profit_margin_budget = coef_budget * (change_in_budget / budget_std)

print(f"Predicted change in profit margin with an increase of $1,000,000 in revenue: {predicted_change_profit_margin_revenue:.2f} units")
print(f"Predicted change in profit margin with an increase of $1,000,000 in budget: {predicted_change_profit_margin_budget:.2f} units")
```

For every increase in $1 million in revenue, profit margin is predicted by the emodel to increase by 4%.

For every increase in $1 million in budget, profit margin is predicted by the model to decrease by 20%.

# Language-Based Analysis of Movie Popularity - VIRGINIA


```python
import pandas as pd
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk

# Ensure necessary resources are downloaded
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')  # Required for lemmatizer languages

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Define custom stop words
default_stop_words = set(stopwords.words('english'))
custom_stop_words = default_stop_words.union({
    'one', 'film', 'two', 'he', 'her', 'his', 'she', 'back', 'do', 'like'
})

# Helper function to clean, remove punctuation, and lemmatize text
def clean_and_lemmatize_text(text):
    if isinstance(text, str):
        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))
        # Remove stop words and lemmatize
        words = [lemmatizer.lemmatize(word.lower()) for word in text.split() if word.lower() not in custom_stop_words]
        return " ".join(words)
    return text

# Data cleaning steps
movie_df_clean = movie_df[movie_df['title'].notna()].copy()  # Remove movies with no title
movie_df_clean = movie_df_clean[movie_df_clean['release_date'].notna()].copy()  # Remove movies with no release date
movie_df_clean['release_date'] = pd.to_datetime(movie_df_clean['release_date'], errors='coerce')  # Convert to datetime
movie_df_clean = movie_df_clean[movie_df_clean['release_date'] < '2030-01-01'].copy()  # Filter by release date
movie_df_clean = movie_df_clean[movie_df_clean['vote_count'] > 0].copy()  # Remove movies with no votes
movie_df_clean = movie_df_clean[movie_df_clean['vote_average'] > 0].copy()  # Remove movies with vote_average == 0
movie_df_clean = movie_df_clean[movie_df_clean['adult'] == False].copy()  # Remove adult movies

# Apply text cleaning to relevant columns
columns_to_clean = ['overview', 'tagline']  # Specify columns to clean
for col in columns_to_clean:
    if col in movie_df_clean.columns:
        movie_df_clean[f'{col}_clean'] = movie_df_clean[col].apply(clean_and_lemmatize_text)

# Display cleaned dataset info
print(movie_df_clean.info())
```

## N-grams analysis (bi-grams and tri-grams)


```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

def analyze_ngrams(text_series, ngram_range):
    """
    Analyze n-grams in a text series and return the top 10 most frequent n-grams.
    """
    vectorizer = CountVectorizer(ngram_range=ngram_range)
    ngram_counts = vectorizer.fit_transform(text_series.dropna())
    ngram_freq = pd.DataFrame(ngram_counts.toarray(), columns=vectorizer.get_feature_names_out()).sum().sort_values(ascending=False).head(10)
    return ngram_freq

# Bigram Analysis for Overview
print("\nTop Bigrams in Popular Movies' Overview:")
popular_overview_bigrams = analyze_ngrams(popular_movies['overview_clean'], ngram_range=(2, 2))
print(popular_overview_bigrams)

print("\nTop Bigrams in Less Popular Movies' Overview:")
less_popular_overview_bigrams = analyze_ngrams(less_popular_movies['overview_clean'], ngram_range=(2, 2))
print(less_popular_overview_bigrams)

# Trigram Analysis for Overview
print("\nTop Trigrams in Popular Movies' Overview:")
popular_overview_trigrams = analyze_ngrams(popular_movies['overview_clean'], ngram_range=(3, 3))
print(popular_overview_trigrams)

print("\nTop Trigrams in Less Popular Movies' Overview:")
less_popular_overview_trigrams = analyze_ngrams(less_popular_movies['overview_clean'], ngram_range=(3, 3))
print(less_popular_overview_trigrams)

# Bigram Analysis for Tagline
print("\nTop Bigrams in Popular Movies' Tagline:")
popular_tagline_bigrams = analyze_ngrams(popular_movies['tagline_clean'], ngram_range=(2, 2))
print(popular_tagline_bigrams)

print("\nTop Bigrams in Less Popular Movies' Tagline:")
less_popular_tagline_bigrams = analyze_ngrams(less_popular_movies['tagline_clean'], ngram_range=(2, 2))
print(less_popular_tagline_bigrams)

# Trigram Analysis for Tagline
print("\nTop Trigrams in Popular Movies' Tagline:")
popular_tagline_trigrams = analyze_ngrams(popular_movies['tagline_clean'], ngram_range=(3, 3))
print(popular_tagline_trigrams)

print("\nTop Trigrams in Less Popular Movies' Tagline:")
less_popular_tagline_trigrams = analyze_ngrams(less_popular_movies['tagline_clean'], ngram_range=(3, 3))
print(less_popular_tagline_trigrams)
```

# Analysis of the audience voting activity - OLEKSANDR.

The main purpose of the numeric analysis of the audience voting activity is to understand what motivates the viewers to vote for a movie and what factors influence the average voting scores. The viewer's voting action can be viewed at a high level as an expression of viewer's engagement, and willingness to share their opinion about the movie.

The dataset sourced from TMDB consisting of 1,127,777 movies will be used for this analysis. Some of the records in the source database are missing important metrics. Therefore, there is the need to use data filtering and cleaning technique to exclude the records that are not suitable for further analysis. In particular, all records with 0 and null values in the vote_count and vote_average columns were filtered out. The resulting subset includes 349,440 movies. Even though the resulting subset is significantly smaller than the initial dataset, 349k of movies is still sufficient volume to reliably identify trends and correlations in the data.

The analysis of voting activity will attempt to answer the following questions:
1. Is there a correlation between vote counts and the average scores?
2. Does the movie budget affect its voting count and average score?
3. Does the movie runtime have an impact on its voting count and average score?
4. What movie genres are most voted? Is it true that the most voted genres are also most highly ranked?
5. Does the movie score count and ranking depend on the production country?

The analisys will be performed and visualized using Python libraries pandas, numpy, statsmodels.api and matplotlib. The results and conclusions are presented below.


```python
# Perform necessary imports
import kagglehub
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib
from matplotlib import pyplot as plt
```

```python
# Download the latest version of the dataset
path2 = kagglehub.dataset_download("asaniczka/tmdb-movies-dataset-2023-930k-movies")

# Specify the correct CSV file name
csv_file2 = os.path.join(path2, 'TMDB_movie_dataset_v11.csv')

# Read the CSV file into a DataFrame
movie_db2 = pd.read_csv(csv_file2)

print("Column names:", movie_db2.columns.tolist())
```

## <span style="color: Orangered;">Is there correlation between vote count and average score? </span>

Correlation between vote_count and vote_average shows if movies with high audience activity (vote_count) are more or less likely to have higher vote average.

We will use *statsmodels* Python library to perform the linear regression analysis. Selecting vote_count and vote_average and dropping all records with 0 values will be sufficient to prepare data for this analysis.

```python
votecount_vs_voteaverage = movie_db2[['vote_count', 'vote_average']].copy()
votecount_vs_voteaverage.info()
```

```python
# Data clean up - remove records with 0 vote count and vote average values
votecount_vs_voteaverage = votecount_vs_voteaverage[votecount_vs_voteaverage['vote_count'] > 0]
votecount_vs_voteaverage = votecount_vs_voteaverage[votecount_vs_voteaverage['vote_average'] > 0]
votecount_vs_voteaverage
```

```python
# For the intercept coefficient, we add new column of 1
votecount_vs_voteaverage['ones'] = np.ones(len(votecount_vs_voteaverage))
# Define X and Y variables for the regression.
Y = votecount_vs_voteaverage['vote_average']
X = votecount_vs_voteaverage[['vote_count','ones']]
result_votecorr = sm.OLS( Y, X ).fit()
result_votecorr.summary()
```

```python
slope, intercept = result_votecorr.params
plt.scatter(votecount_vs_voteaverage['vote_count'], votecount_vs_voteaverage['vote_average'])
plt.plot(votecount_vs_voteaverage['vote_count'], intercept + slope*votecount_vs_voteaverage['vote_average'], 'k' )
plt.title("Vote count vs. vote average")
plt.xlabel("Vote count")
plt.ylabel("Vote score average")
```

The diagram above shows a scatterplot and a linear regression line for Vote Count vs. Vote Average correlation.

The scatterplot clearly shows that the higher the vote count for particular movie is, the lower is the chance that this movie will receive low average score. None of the movies with vote count exceeding 10k received average score lower than 5.6.

Low rankings (below 4) can only be received by those movies that have low vote counts (1k and less).

The linear regression of average vote scores from vote counts is roughly parallel to X axis and is positioned at the level y=6.09. Looking at the scatterplot diagram, we would expect that the linear regression line would be trending more upward. However, this is not the case because the density of the points in the left sector of the diagram is significantly higher and has the determining influence on the incline angle of the regression line. Points located in the middle and top right sector of the scatterplot are very scarce and they have very low influence on the regression line.

Therefore, we can conclude that there is a negligible regression of average vote scores from vote counts.

## <span style="color: Orangered;">Global vote average </span>

Global Vote Average will be calculated as weighted average of the movie vote scores scaled by vote count per each movie.

```python
def weighted_avg(df, score, count):
    movieScore = df[score]
    voteCount = df[count]
    return (movieScore * voteCount).sum() / voteCount.sum()

global_vote_average = weighted_avg(votecount_vs_voteaverage, 'vote_average', 'vote_count')
global_vote_average
```

The result tells us that when taking into account all the votes given to all the movies in the dataset, the average score given by one vote is **~6.83**. The practical meaning of this metric is that viewers on average tend to give high scores (rather than low scores) when they vote for the movies.

## <span style="color: Orangered;">Does the audience engagement depend on the movie budget? </span>

To answer this question we will group the movies according to their budgets and for each group we will identify the average vote count per movie.

For this analysis, we will need to take into consideration the column *budget* that has a lot of values requiring cleaning. For example, there are budgets exceeding 500m and budgets below 100, that are not realistic values. To make sure the analysis is performed on the accurate data, we will set maximum (450,000,000) and minimum (2,000) budget thresholds and remove all the records that do not fit into this range.

The size of dataset after cleanup is pretty low at 22,061 records. Although, in this case, it is preferable to have small dataset of reliable data rather than large set with questionable budget values.

The distribution of the budget amount values in the resulting subset is not even. Therefore, we will use the *pd.qcut* function for defining 4 budget amount range groups to make sure that each group contains equal amount of movies in it.

```python
vcBudget = movie_db2[['vote_count', 'vote_average', 'budget', 'status']].copy()
vcBudget.info()
```

```python
#Data clean up - remove records without vote counts, vote average. Drop records with unrealistic budget values.
vcBudget = vcBudget[
    (vcBudget['vote_count'] > 0) &
    (vcBudget['vote_average'] > 0) &
    (vcBudget['budget'] > 2000)  &
    (vcBudget['budget'] <= 450_000_000) &
    (vcBudget['status'] == 'Released')
]
vcBudget
```

```python
# Define budget amount ranges to split movies in 4 groups of equal size.
pd.qcut(vcBudget['budget'],4,precision=0).value_counts()
```

```python
# Function to group by budget value
def budgetToRange(a):
    if a < 150000.0:
        return '2k-150k'
    elif a >= 150000.0 and a < 2000000.0:
        return '150k-2m'
    elif a >= 2000000.0 and a < 10000000.0:
        return '2m-10m'
    else:
        return 'Over 10m'


#Group the movies by budget range
vcBudget['budgetRange'] = vcBudget['budget'].map(budgetToRange)
groupedVcBudget = vcBudget['vote_count'].groupby(vcBudget['budgetRange'])
totalVoteCountByBudget = groupedVcBudget.sum()
totalVoteCountByBudget = totalVoteCountByBudget.reindex(['2k-150k', '150k-2m', '2m-10m', 'Over 10m'])
totalVoteCountByBudget
```

```python
#Calculate average vote count per group
avgVoteCountByBudget = groupedVcBudget.mean()
avgVoteCountByBudget = avgVoteCountByBudget.reindex(['2k-150k', '150k-2m', '2m-10m', 'Over 10m'])

# Plot the histogram of total vote count for each budget group
plt.style.use('seaborn-v0_8-deep')
plt.rcParams["figure.figsize"] = (6,6)
bar_plot = totalVoteCountByBudget.plot(x = 'budget range',  y = 'num of votes', kind='bar', color = 'darkolivegreen', width=0.9)
plt.gca().yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, p: format(int(y), ',')))
plt.title("Total vote count per movie budget group")
plt.ylabel("Total number of votes")
bar_plot.bar_label(bar_plot.containers[0], fmt='{:,.0f}')
```

```python
# Plot the histogram of average vote count per movie for each budget group
bar_plot2 = avgVoteCountByBudget.plot(x = 'Budget range',  y = 'Avg votes per movie', kind='bar', color = 'darkolivegreen', width=0.9)
plt.gca().yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y, p: format(int(y), ',')))
plt.title("Average vote count per movie in each budget group")
plt.ylabel("Average number of votes per movie")
bar_plot2.bar_label(bar_plot2.containers[0], fmt='{:,.0f}')
```

All the movies were split in 4 equal sized groups depending on their budget amount. The size of one group is around 5,500 movies.

The first histogram visualizes total count of votes given to movies in each budget group. Majority of votes are being received by the movies with budgets over 10m (13.1m votes). The second histogram presents the average vote count per movie for each budget range group. The results are consistent with the first diagram.

- The lowest vote count is given to movies in the lowest budget group (2k-150k) - 23 votes per movie.

## <span style="color: Orangered;">What movie genres are most voted? </span>

In this section we will identify the relationship between movie genre and its vote scores and vote counts.

There are 19 movie genres. One movie can belong to several movie genres at the same time, in which case value in the column 'genre' will contain a CSV list with applicable genre names (for example, 'Action, Science Fiction, Adventure'). In order to perform the analysis, we will decompose the records that have multiple genre names in the 'genre' column. One record with multiple genre names will be transformed into multiple records where each record will have only one genre name. For example, if a movie with id 123 belongs to genre 'Action, Adventure, Science Fiction', then as a result of transformation we will have 3 records: 'id 123 genre Action', 'id 123 genre Adventure', 'id 123 genre Science Fiction'. Normalization process increases number of records in dataset from 290k to 504k. All records will be grouped by single genre name and vote counts and vote scores will be calculated on each group for further comparison.

```python
import kagglehub
import os
import pandas as pd

# Read Movie CSV file
def read_movies(csv):
    try:
        # Read the CSV file
        df = pd.read_csv(csv)
        return df

    except FileNotFoundError:
        print("Error: The file {} was not found.".format(csv))
    except Exception as e:
        print("Error occurred: {}".format(e))

# Download the latest version of the dataset
path = kagglehub.dataset_download("asaniczka/tmdb-movies-dataset-2023-930k-movies")

# Specify the correct CSV file name
csv_file = os.path.join(path, 'TMDB_movie_dataset_v11.csv')

# Load TMDB Movies file
#csv_file = 'TMDB_movie_dataset_v11.csv'
movies = read_movies(csv_file)

# Display basic information
print(movies.info())

# Display movies after date concatenation
movies.head()
```

```python
# Data cleanup
import numpy as np

# Remove movies with no title
movies_clean = movies[movies['title'].notna()]

# Remove movies with no release data
movies_clean = movies_clean[movies_clean['release_date'].notna()]

# Remove movies where the release date is on or after 2030
movies_clean = movies_clean[movies_clean['release_date'].between('1915','2024')]

# Remove Adult movies from the analysis
movies_clean = movies_clean[~movies_clean['adult']]

# Remove movies with Revenue of 0 and Vote Count < 10
movies_clean = movies_clean[~(movies_clean['revenue'] == 0)]
movies_clean = movies_clean[~(movies_clean['budget'] == 0)]
movies_clean = movies_clean[~(movies_clean['vote_count'] < 10)]

# Removed movies with NaN or infinite values in Production Cost or Revenue
movies_clean = movies_clean.replace([np.inf, -np.inf], np.nan).dropna(subset=['budget', 'revenue'])

# Specific bad data removal
movies_clean = movies_clean[~(movies_clean['id'].isin([1224207,1270893, 1029460]))]

#movies_clean.sort_values(by='revenue', ascending=False, inplace=False).head()
print(movies_clean.info())
```

```python

#%%
# Extrapolating movie genres into individual values and only looking at the first 2 Genres (0, 1)

genres = movies_clean['genres'].str.split(', ', expand=True)
genres.columns = ['genre_' + str(col) for col in genres.columns]
movies_clean = pd.concat([movies_clean, genres['genre_0'], genres['genre_1']], axis=1)
movies_clean.head(25)
```

```python
# Drop non-relevant data
movies_clean.drop([
    'adult',
    'backdrop_path',
    'homepage',
    'imdb_id',
    'overview',
    'poster_path',
    'tagline',
    'genres',
    'production_companies',
    'production_countries',
    'keywords'],
    axis=1, inplace=True)

movies_clean.info()
```

```python

# Breaking down movies by decades}
movies_clean['decade'] = (pd.to_datetime(movies_clean['release_date']).dt.year // 10) * 10
```

```python

# Adjust the revenue and budget dollar amounts to 2024 dollar amounts
cpi_data = {
    1850: 8.3, 1860: 9.1, 1870: 12.2, 1880: 10.2, 1890: 8.5, 1900: 9.7,
    1910: 10.5, 1920: 20.0, 1930: 17.1, 1940: 16.7, 1950: 24.1, 1960: 29.6,
    1970: 38.8, 1980: 82.4, 1990: 130.7, 2000: 172.2, 2010: 218.1, 2020: 258.8, 2024: 285.0
}

# Reduce $ by a Billion to manage number lengths
reduction = 1000000000

# Function to adjust for inflation to 2024 dollars
def adjust_for_inflation(cpi_data, amount, from_year, to_year=2024):
    if from_year in cpi_data and to_year in cpi_data:
        return round((amount * (cpi_data[to_year] / cpi_data[from_year])) / reduction, 3)
    return amount

# Calculate adjusted dollar amounts for revenue and budget add as a new column
movies_clean['adjusted_revenue'] = movies_clean.apply(lambda row: adjust_for_inflation(cpi_data, row['revenue'], row['decade']), axis=1)
movies_clean['adjusted_budget'] = movies_clean.apply(lambda row: adjust_for_inflation(cpi_data, row['budget'], row['decade']), axis=1)
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Preparing data for the stacked histograms
stacked_revenue_data = movies_clean.pivot_table(
    values='adjusted_revenue',
    index='decade',
    columns='genre_0',
    aggfunc='sum',
    fill_value=0
)

stacked_budget_data = movies_clean.pivot_table(
    values='adjusted_budget',
    index='decade',
    columns='genre_0',
    aggfunc='sum',
    fill_value=0
)

# Calculate total budget and revenue for each decade
total_budget_by_decade = stacked_budget_data.sum(axis=1)
total_revenue_by_decade = stacked_revenue_data.sum(axis=1)

# Max values to dynamically determine the scale of the plots to easily compare budget vs revenue
max_value = max(total_budget_by_decade.values.max(), total_revenue_by_decade.values.max())+10

# Dynamically generate unique colors
n_colors = len(stacked_revenue_data.columns)
colors = plt.get_cmap('tab20', n_colors)
color_map = {genre: colors(i) for i, genre in enumerate(stacked_revenue_data.columns)}

# Plotting revenue and budget as separate subplots with consistent genre colors
fig, axes = plt.subplots(2, 1, figsize=(16, 14), sharex=True)

# Plot Budget
stacked_budget_data.plot(
    kind='bar', stacked=True, ax=axes[0], color=[color_map[genre] for genre in stacked_budget_data.columns]
)
axes[0].plot(range(len(total_revenue_by_decade)), total_revenue_by_decade.values,
             marker='o', color='black', linewidth=1, label='Total Revenue for Decade')
axes[0].set_title('Total Adjusted Budget ($B) for 2024 by Genre Across Decades', fontsize=24)
axes[0].set_xlabel('Decade', fontsize=12)
axes[0].set_ylabel('Total Budget (in $B)', fontsize=12)
axes[0].legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper center')
axes[0].set_ylim(0, max_value)

# Plot Revenue
stacked_revenue_data.plot(
    kind='bar', stacked=True, ax=axes[1], color=[color_map[genre] for genre in stacked_revenue_data.columns]
)
axes[1].plot(range(len(total_budget_by_decade)), total_budget_by_decade.values,
             marker='o', color='black', linewidth=1, label='Total Budget for Decade')
axes[1].set_title('Total Adjusted Revenue ($B) for 2024 by Genre Across Decades', fontsize=24)
axes[1].set_xlabel('Decade', fontsize=12)
axes[1].set_ylabel('Total Revenue (in $B)', fontsize=12)
axes[1].legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper center')
axes[1].set_ylim(0, max_value)

plt.tight_layout()
plt.show()
```
